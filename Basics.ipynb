```python
import csv
import datetime
```

```python
f = open('guns.csv','r')
reader = csv.reader(f)
data_list = list(reader)
```

```python
headers = data_list[0]
data = data_list[1:]
#headers
#data
```

```python
years = [data[1] for data in data]

years_count = dict()

for row in years:
    if(row in years_count):
        years_count[row] = years_count[row] + 1
    else:
        years_count[row] = 1
```
